```python
# Import necessary modules

import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector, plot_histogram
```

```python
# Question 5 (a)
# A circuit that swaps the state of two qubits

qc = QuantumCircuit(2)
qc.swap(0, 1)
qc.draw(output='mpl')
```

```python
# Let's see the state of the qubits after the swap
sv = Statevector.from_label('01')
sv = sv.evolve(qc)
plot_state_qsphere(sv)
```

```python
# Question 5 (b)
# A circuit that takes a computational basis to a Hadamard basis
qc = QuantumCircuit(1)
qc.h(0)
qc.draw(output='mpl')
```

```python
# Let's see the state of the qubit after the Hadamard gate
sv = Statevector.from_label('0')
sv = sv.evolve(qc)
plot_bloch_multivector(sv)
```

```python
# Question 5 (c)
# Create the GHZ State
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.cx(0, 2)
qc.draw(output='mpl')
```

```python
# Let's see the state of the qubits after the GHZ state
sv = Statevector.from_label('000')
sv = sv.evolve(qc)
plot_state_qsphere(sv)
```

```python
# Question 5 (c) - second part
# Create the W State
qc = QuantumCircuit(3)
qc.ry(1.9106332362490184, 0)
qc.ch(0, 1)
qc.cx(1, 2)
qc.cx(0, 1)
qc.x(0)
qc.draw(output='mpl')
```

```python
# Let's see the state of the qubits after the W state
sv = Statevector.from_label('000')
sv = sv.evolve(qc)
plot_state_qsphere(sv)
```

```python
# Question 5 (d)
# A circuit that decrements a three bit number by 1 and 
# stores the result in the same qubits that are used for the input
qc = QuantumCircuit(3)
qc.x(0)
qc.cx(0, 1)
qc.ccx(0, 1, 2)
qc.draw(output='mpl')
```

```python
sv = Statevector.from_label('100')
sv = sv.evolve(qc)
sv.draw('latex')
```

```python
labels = [f'{i:03b}' for i in range(8)]
labels = labels[1:]
for i, label in enumerate(labels):
    sv = Statevector.from_label(label)
    sv = sv.evolve(qc)
    sv.draw('latex', prefix=f'{i+1}. ')
```
